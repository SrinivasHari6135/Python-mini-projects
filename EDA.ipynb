```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('students.csv')
print(df)
print(df.describe())

df['math'].hist(bins=10)
plt.title('Math Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df, x='math', y='science')
plt.title('Math vs Science')
plt.show()

# Correlation
print(df[['math','science','english']].corr())
```

```python
# Box plot per subject
df[['math','science','english']].plot(kind='box')
plt.title('Subject Score Distribution')
plt.ylabel('Score')
plt.show()
```

```python
# Grade distribution
grade_counts = df['grade'].value_counts().sort_index()
grade_counts.plot(kind='bar', color='teal')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```
